##Numpy e operações básicas

```python
import numpy as np

# Array a partir de uma lista
array1 = np.array([1, 2, 3, 4, 5])
print("Array1:", array1)

# Array de zeros
array2 = np.zeros((2, 3))
print("Array2 (zeros):\n", array2)

# Array de uns
array3 = np.ones((3, 3))
print("Array3 (ones):\n", array3)

# Array com uma sequência de números
array4 = np.arange(0, 10, 2)
print("Array4 (arange):", array4)
```

```python
# Operações aritméticas
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Soma:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
```

```python
#Diferença entre somar listas e somar arrays
lista1 = [4, 5, 6]
lista2 = [1, 2, 3]
lista1 + lista2
```

```python
# Funções universais
array = np.array([1, 2, 3, 4, 5])
print("Soma dos elementos:", np.sum(array))
print("Média dos elementos:", np.mean(array))
print("Máximo dos elementos:", np.max(array))
print("Desvio padrão dos elementos:", np.std(array))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount("/content/drive")
dados = pd.read_csv("/content/drive/MyDrive/base/venda1.csv")
df_dados = pd.DataFrame(dados)
```

**Informações do dataset:**
Este conjunto de dados fornece uma visão geral abrangente das transações de vendas online em diferentes categorias de produtos. Cada linha representa uma única transação com informações detalhadas como ID do pedido, data, categoria, nome do produto, quantidade vendida, preço unitário, preço total, região e forma de pagamento.

```python
df_dados.replace(np.nan, 0, inplace=True)
df_dados.head()
```

```python
df_dados["Unit Price"] = df_dados["Unit Price"].str.replace(',', '.')
df_dados["Unit Price"] = df_dados["Unit Price"].astype(float)
df_dados.head()
```

```python
df_dados.info()
```

```python
df_dados["Total Revenue"] = df_dados["Total Revenue"].str.replace(',', '.').astype(float)
df_dados.head()
```

```python
vendas_por_categoria = df_dados.groupby('Product Category')['Units Sold'].sum()
vendas_por_categoria
```

```python
vendas_por_categoria.plot(kind='bar')
plt.title('Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
vendas_por_categoria = df_dados.groupby('Product Category')['Units Sold'].sum().reset_index()
fig = px.bar(vendas_por_categoria, x='Product Category', y='Units Sold', title='Vendas por Categoria')
fig.show()
```

```python
vendas_por_regiao = df_dados.groupby('Region')['Units Sold'].sum()
vendas_por_regiao.plot(kind='bar')
plt.title('Vendas por Região')
plt.xlabel('Região')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
vendas_por_regiao = df_dados.groupby('Region')['Units Sold'].sum().reset_index()
fig = px.bar(vendas_por_regiao, x='Region', y='Units Sold', title='Vendas por Região')
fig.show()
```

```python
vendas_por_metodo = df_dados.groupby('Payment Method')['Units Sold'].sum()
vendas_por_metodo.plot(kind='bar')
plt.title('Vendas por Método de Pagamento')
plt.xlabel('Método de Pagamento')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
df_dados['Date'] = pd.to_datetime(df_dados['Date'])
vendas_por_tempo = df_dados.groupby('Date')['Units Sold'].sum()
vendas_por_tempo.plot(kind='line')
plt.title('Tendência das Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
df_dados['Date'] = pd.to_datetime(df_dados['Date'])
vendas_por_tempo = df_dados.groupby('Date')['Units Sold'].sum().reset_index()
fig = px.line(vendas_por_tempo, x='Date', y='Units Sold', title='Tendência das Vendas ao Longo do Tempo')
fig.show()
```

```python
sns.scatterplot(data=df_dados, x='Payment Method', y='Unit Price')
plt.title('Correlação entre Método de Pagamento e Quantidade Vendida')
plt.xlabel('Método de Pagamento')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
fig = px.scatter(df_dados, x='Payment Method', y='Unit Price', title='Correlação entre Método de Pagamento e Quantidade Vendida')
fig.show()
```

```python
vendas_cat_reg = df_dados.groupby(['Product Category', 'Region'])['Units Sold'].sum().unstack()
vendas_cat_reg.plot(kind='bar', stacked=True)
plt.title('Vendas por Categoria e Região')
plt.xlabel('Categoria')
plt.ylabel('Quantidade Vendida')
plt.show()
```
